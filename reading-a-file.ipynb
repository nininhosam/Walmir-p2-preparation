# PDF

```python
import pdfplumber

def load_pdf(filepath):
    print(f"Lendo PDF: {filepath}")
    with pdfplumber.open(filepath) as pdf:
        text = ""
        for page in pdf.pages[:2]:  # preview das 2 primeiras páginas
            text += page.extract_text() or ""

    print("\nPrévia do PDF:")
    print(text[:800])
    return text


# Exemplo:
load_pdf("./documents/sample.pdf")
```

# Json

```python
import json

def load_json(filepath):
    print(f"Lendo JSON: {filepath}")
    with open(filepath, "r", encoding="utf-8") as f:
        data = json.load(f)

    print("\nPrévia do JSON:")
    print(str(data)[:800])
    return data


# Exemplo:
load_json("./documents/sample.json")
```

# XLSX

```python
import pandas as pd

def load_xlsx(filepath):
    print(f"Lendo XLSX: {filepath}")
    df = pd.read_excel(filepath)

    print("\nPrévia do XLSX:")
    display(df.head())
    return df


# Exemplo:
load_xlsx("./documents/sample.xlsx")
```

# TXT

```python
from pathlib import Path

def load_txt(filepath):
    print(f"Lendo TXT: {filepath}")
    text = Path(filepath).read_text(encoding="utf-8")

    print("\nPrévia do TXT:")
    print(text[:800])
    return text


# Exemplo:
load_txt("./documents/sample.txt")
```

# DOCX

```python
import docx

def load_docx(filepath):
    print(f"Lendo DOCX: {filepath}")
    doc = docx.Document(filepath)

    full_text = "\n".join([p.text for p in doc.paragraphs])

    print("\nPrévia do DOCX:")
    print(full_text[:800])
    return full_text


# Exemplo:
load_docx("./documents/sample.docx")
```
